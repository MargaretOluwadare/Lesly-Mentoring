```python
# Email sending function
def send_email(recipient, subject, body):
    try:
        sender_email = "your_email@example.com"
        sender_password = "your_password"

        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient
        msg['Subject'] = subject
        msg.attach(MIMEText(body, 'plain'))

        with smtplib.SMTP('smtp.gmail.com', 587) as server:
            server.starttls()
            server.login(sender_email, sender_password)
            server.send_message(msg)

        print("Email sent successfully!")
    except Exception as e:
        print(f"Failed to send email: {e}")

# Task scheduling function
def schedule_task(task, time):
    try:
        task_time = datetime.datetime.strptime(time, "%Y-%m-%d %H:%M")
        now = datetime.datetime.now()

        if task_time > now:
            delay = (task_time - now).total_seconds()
            print(f"Task scheduled: '{task}' at {time}")
            os.system(f"echo '{task}' | at {time}")
        else:
            print("Scheduled time must be in the future.")
    except Exception as e:
        print(f"Failed to schedule task: {e}")

# Voice command function
def listen_command():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source)
            command = recognizer.recognize_google(audio)
            print(f"You said: {command}")
            return command.lower()
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")
        except sr.RequestError as e:
            print(f"Could not request results; {e}")

# Voice response function
def respond_with_voice(response):
    engine = pyttsx3.init()
    engine.say(response)
    engine.runAndWait()

# Main function
def main():
    print("Hello! I am your AI personal assistant.")
    respond_with_voice("Hello! I am your AI personal assistant.")

    while True:
        command = listen_command()

        if "email" in command:
            respond_with_voice("Who should I send the email to?")
            recipient = input("Enter recipient email: ")
            respond_with_voice("What should be the subject?")
            subject = input("Enter subject: ")
            respond_with_voice("What should I say in the email?")
            body = input("Enter email body: ")
            send_email(recipient, subject, body)

        elif "schedule" in command:
            respond_with_voice("What task should I schedule?")
            task = input("Enter task: ")
            respond_with_voice("When should I schedule it? Please say the time in YYYY-MM-DD HH:MM format.")
            time = input("Enter time: ")
            schedule_task(task, time)

        elif "exit" in command:
            respond_with_voice("Goodbye!")
            print("Goodbye!")
            break

        else:
            respond_with_voice("Sorry, I didn't understand that command.")

if __name__ == "__main__":
    main()
```

```python
import smtplib
import schedule
import time
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from datetime import datetime
import re

# Define a simple helper function to handle the email sending
def send_email(subject, body, to_email):
    sender_email = "your_email@gmail.com"
    sender_password = "your_email_password"  # Use an app password if 2FA is enabled

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = to_email
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    try:
        with smtplib.SMTP('smtp.gmail.com', 587) as server:
            server.starttls()  # Secure connection
            server.login(sender_email, sender_password)
            text = msg.as_string()
            server.sendmail(sender_email, to_email, text)
        print("Email sent successfully!")
    except Exception as e:
        print(f"Error sending email: {e}")

# Define a simple function to schedule appointments
def schedule_appointment(time_str, appointment_details):
    # This function uses the schedule library to schedule tasks
    def job():
        print(f"Reminder: {appointment_details}")
    
    schedule_time = datetime.strptime(time_str, '%Y-%m-%d %H:%M')
    schedule.every().day.at(schedule_time.strftime("%H:%M")).do(job)
    
    print(f"Appointment scheduled for {time_str} with details: {appointment_details}")

# Define regex-based query handler
def handle_query_with_regex(query):
    # Define possible actions using regex patterns
    action_patterns = {
        r"(schedule|appointment)": "schedule_appointment",
        r"(send|email)": "send_email",
        r"(reminder)": "set_reminder",
        r"(meeting)": "schedule_appointment"
    }

    # Search for patterns in the query
    for pattern, action in action_patterns.items():
        if re.search(pattern, query, re.IGNORECASE):
            # Call the function based on the matched pattern
            if action == "schedule_appointment":
                try:
                    # Extract date and time info from the query (assuming it's well-formed like "schedule appointment for 2024-12-12 14:00")
                    time_str = query.split("for")[1].strip()
                    appointment_details = "Doctor's Appointment"
                    schedule_appointment(time_str, appointment_details)
                    return f"Appointment scheduled for {time_str}."
                except Exception as e:
                    return f"Error scheduling appointment: {e}"
            elif action == "send_email":
                subject = "Meeting Reminder"
                body = "This is a reminder for the meeting scheduled tomorrow."
                to_email = "recipient_email@example.com"  # Replace with recipient email
                send_email(subject, body, to_email)
                return "Email sent successfully!"
            elif action == "set_reminder":
                # Placeholder for reminder functionality (could be enhanced further)
                return "Reminder set!"
    
    return "Sorry, I didn't understand that."

# Main loop to listen for text commands
def main():
    while True:
        query = input("How can I assist you today? (Type 'exit' to quit): ").lower()

        if query == 'exit':
            print("Goodbye!")
            break

        response = handle_query_with_regex(query)
        print(response)

        # Run scheduled tasks in the background
        while True:
            schedule.run_pending()
            time.sleep(1)

if __name__ == "__main__":
    main()
```
